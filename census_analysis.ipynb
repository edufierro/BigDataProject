# Crime data

```python
data_crime = pd.read_csv("../NYPD_Complaint_Data_Historic.csv")
```

```python
boroughs = list(np.unique(list(data_crime['BORO_NM'])))
boroughs = [x for x in boroughs if str(x) != 'nan']
boroughs

df_entire = data_crime[['CMPLNT_NUM', 'CMPLNT_FR_DT', 'BORO_NM']]
df_entire = df_entire.dropna()
df_entire['CMPLNT_FR_DT'] = [int(x.split("/")[2]) for x in list(df_entire['CMPLNT_FR_DT'])]
df_entire['CMPLNT_FR_DT'] = df_entire['CMPLNT_FR_DT'].astype(int)
df_entire = df_entire[df_entire['CMPLNT_FR_DT'] >= 2005 ]
df_entire = df_entire[df_entire['CMPLNT_FR_DT'] <= 2015]
```

```python
poverty_data_borough.shape
```

```python
df_subset = df_entire.reset_index(drop = True).groupby(['CMPLNT_FR_DT', 'BORO_NM']).count().reset_index()
poverty_data_borough['Crime count'] = list(df_subset['CMPLNT_NUM'])
```

## Save data to disk

```python
a = poverty_data_borough['Total; Estimate; Population for whom poverty status is determined - EDUCATIONAL ATTAINMENT - Population 25 years and over - Less than high school graduate']
b = poverty_data_borough['Total; Estimate; Population for whom poverty status is determined - EDUCATIONAL ATTAINMENT - Population 25 years and over - High school graduate (includes equivalency)']
c = poverty_data_borough['Total; Estimate; Population for whom poverty status is determined - EDUCATIONAL ATTAINMENT - Population 25 years and over - Some college, associate\'s degree']

poverty_data_borough['Percent; Estimate; Population for whom poverty status is determined - EDUCATIONAL ATTAINMENT - Population 25 years and over - population percent with only high school'] = a/(a+b+c)*100
```

```python
population_data_census.to_csv('population_data_census.csv')
poverty_data_borough.to_csv('poverty_data_borough.csv')
```
