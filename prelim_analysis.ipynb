# OUTLIER DETECTION:

* COMPLAINT DATES: 

 - Lots of NA here. Crime date is not being reported. Also, 6 complaints were allegedly reported during the                        year 1015. This is clearly an invalid number. And these records should be removed (or be imputed with 2015)

* COMPLAINT TIMINGS:

 - It looks like the timings reported are in 24 hour format. But still one where the complaint time registered is 24:00:00. If the time is 24 hur format, this should be converted to 00:00:00. (total missing = 48)
 
 
* ENDING DATE OF COMPLAINT:
 
 - There is one date where year = 2090. This is obviously an invalid number and one possible imputation could be 1990. 
 - Also, there are lots of NA values (precisely = 1,391,478) . This should probably be because most of the cases aren't solved yet.
 - Also, one weird fact - In 31 cases, ending time of the complaint is before the complaint was filed (HIGHLY IMPOSSIBLE!). The table is shown below
 
 
* ENDING TIME OF COMPLAINTS:

 - Similar to Complaint timings variables. It has one value with 24:00:00, which needs to be replaced by 00:00:00. Also, there are lots of NA values (precisely = 1,387,785, which is around similar to the number of missing complaint dates). It is again because most of the cases aren't solved yet.

 - No anomaly here. Also, there are no missing values. This makes sense as if we had missing values in reporting time, then these cases wouldn't have been reported and wouldn't have been in this complaints csv file.

## To check whether cases where reporting time is >2 ahead of the occurence date
## to check whether closing date is before the opening date


```python

indices = []

dates1 = list(data['CMPLNT_FR_DT'])
dates2 = list(data['RPT_DT'])

for i in range(0,data.shape[0]):
    
    if ((type(dates1[i]) != float) and (type(dates2[i]) != float)):
        day1 = dates1[i][3:5]
        month1 = dates1[i][0:2]
        year1 = dates1[i][6:]

        day2 = dates2[i][3:5]
        month2 = dates2[i][0:2]
        year2 = dates2[i][6:]

        if(int(year2) > int(year1)+2):
            indices.append(i)

print(len(indices))
data.loc[indices,['CMPLNT_FR_DT', 'RPT_DT']]
```

## # to check whether closing date is before the opening date

```python
indices = []

dates1 = list(data['CMPLNT_FR_DT'])
dates2 = list(data['CMPLNT_TO_DT'])
    
for i in range(0,data.shape[0]):
    
    if ((type(dates1[i]) != float) and (type(dates2[i]) != float)):
        day1 = dates1[i][3:5]
        month1 = dates1[i][0:2]
        year1 = dates1[i][6:]

        day2 = dates2[i][3:5]
        month2 = dates2[i][0:2]
        year2 = dates2[i][6:]

        if(year2 < year1):
            indices.append(i)
        elif(year2 == year1 and month2 < month1):
            indices.append(i)
        elif(year2 == year1 and month2 == month1 and day2 < day1):
            indices.append(i)
```

```python
data.loc[indices,['CMPLNT_FR_DT', 'CMPLNT_TO_DT']]
```

## Code to see the realtions between KY_CD AND PD_CD

```python
code_map_dt = data[['KY_CD', 'PD_CD']]
code_map = []

for i in range(0, data.shape[0]):
    key, val = code_map_dt['PD_CD'][i], code_map_dt['KY_CD'][i]
    code_map.append({'key':key, 'val':val})
```

```python
code_map.sort(key=lambda x:x['key'])
```

```python
curr_key = code_map[0]['key']
list_of_curr_key_val = [code_map[0]['val']]
final_list = []

for i in range(1, len(code_map)):
    key = code_map[i]['key']
    val = code_map[i]['val']
    
    if np.isnan(val) != True:
        if key == curr_key:
            list_of_curr_key_val.append(val)
        else:
            final_list.append({curr_key:list(np.unique(list_of_curr_key_val))})
            curr_key = key
            list_of_curr_key_val = [val]
        
```

```python
final_list
```
